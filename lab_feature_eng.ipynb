```python
#Libraries
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.neighbors import KNeighborsRegressor
```

```python
spaceship = pd.read_csv("https://raw.githubusercontent.com/data-bootcamp-v4/data/main/spaceship_titanic.csv")
spaceship.head()
```

**Check the shape of your data**

```python
spaceship.shape
```

**Check for data types**

```python
spaceship.dtypes
```

**Check for missing values**

```python
spaceship.isna().sum()
```

There are multiple strategies to handle missing data

- Removing all rows or all columns containing missing data.
- Filling all missing values with a value (mean in continuos or mode in categorical for example).
- Filling all missing values with an algorithm.

For this exercise, because we have such low amount of null values, we will drop rows containing any missing value.

```python
spaceship = spaceship.dropna()
```

- **Cabin** is too granular - transform it in order to obtain {'A', 'B', 'C', 'D', 'E', 'F', 'G', 'T'}

```python
spaceship["Cabin"].head()

spaceship["Cabin"] = spaceship["Cabin"].str[0]
spaceship["Cabin"].head()
```

- Drop PassengerId and Name

```python
columns_to_drop = [col for col in ["PassengerId", "Name"] if col in spaceship.columns]
if columns_to_drop:
    spaceship = spaceship.drop(columns_to_drop, axis=1)
spaceship.head()
```

- For non-numerical columns, do dummies.

```python
spaceship = pd.get_dummies(spaceship)
spaceship.head()
```

**Perform Train Test Split**

```python
X = spaceship.drop('Transported', axis=1)
y = spaceship['Transported']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.20, random_state=0)
```

**Model Selection**

In this exercise we will be using **KNN** as our predictive model.

```python
knn = KNeighborsRegressor(n_neighbors=10)
```

- Evaluate your model's performance. Comment it

```python
knn.fit(X_train, y_train)
```

```python
print(f"The R2 of the model on the TEST set is: {knn.score(X_test, y_test): .2f}")
```

An R² of 0.40 means that approximately 40% of the variance in whether a passenger is 'Transported' can be explained by the model's features.

Below is me testing things out.

Evaluating with Classification Metrics

Since the target variable `Transported` is binary (True/False), this is a classification problem. While `KNeighborsRegressor` attempts to predict a continuous value (often between 0 and 1 in this context), we can convert these predictions into binary classes (e.g., if the prediction is >= 0.5, classify as True; otherwise, classify as False) and then use standard classification metrics.

```python
from sklearn.metrics import accuracy_score, precision_score, recall_score, f1_score

# Make predictions on the test set
y_pred_proba = knn.predict(X_test)

# Convert probabilities to binary predictions (0 or 1)
# A common threshold is 0.5, but this can be tuned.
y_pred = (y_pred_proba >= 0.5).astype(bool)

# Calculate classification metrics
accuracy = accuracy_score(y_test, y_pred)
precision = precision_score(y_test, y_pred)
recall = recall_score(y_test, y_pred)
f1 = f1_score(y_test, y_pred)

print(f"Accuracy: {accuracy:.2f}")
print(f"Precision: {precision:.2f}")
print(f"Recall: {recall:.2f}")
print(f"F1-Score: {f1:.2f}")
```

These scores give us a much clearer picture of the model's performance on this classification task compared to the R² score.


* An Accuracy of 0.78 means that the model correctly predicted whether a passenger was 'Transported' or not about 78% of the time.
* A Precision of 0.76 indicates that when the model predicts a passenger was 'Transported', it was correct 76% of the time.
* A Recall of 0.83 suggests that the model successfully identified 83% of all passengers who were actually 'Transported'.
* The F1-Score of 0.79 is a balanced measure, considering both precision and recall. It's a good overall indicator of the model's performance.
